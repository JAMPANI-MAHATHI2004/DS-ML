```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme() # sets default theme - whitegrid, darkdark, white, dark, ticks
```

```python
sns.get_dataset_names() # lists available datasets
```

```python
tips = sns.load_dataset("tips")
```

```python
print(type(tips))
print(tips.head())
```

```python
sns.set_theme(style="darkgrid")

# relational plot
sns.relplot(
    data= tips,
    x= "total_bill",
    y= "tip",
    hue= "smoker",
    size = "size",
    style="smoker"
)
```

```python
x_vals = [1,2,4,5,6,7,8,9,10]
y_sq = [i**2 for i in x_vals]


sns.set_theme(style="ticks")
sns.relplot(
    x= x_vals,
    y= y_sq,
    kind = "line"
)
```

```python
"""
Relational plots:
1. scatter plot
2. line plot
"""
```

```python
# scatter plot
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "time"
)
```

```python
# line plot
sns.lineplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
```

```python
# line plot
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
plt.show()

# the shade around the line represents the confidence interva (ci) mean +/- 95%(ci)
```

```python
"""Categorical plots:
1. Bar plots
2. Box plots"""
```

```python
# Bar plot
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.show()

# Box plot
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)
plt.show()
```

```python
# Distribution plots -> histogram
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)

print(penguins.head())
plt.show()
```

```python
"""Matrix plots:
Heatmap - plot rectangular data as a color - encoded matrix
"""
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
print(flights_pivot)

sns.heatmap(
    flights_pivot,
    cmap="coolwarm", annot=True, fmt="d"
)
plt.title("Heatmap for Passengers")
plt.tight_layout()
plt.show()
```

```python
# Recommended approach
fig, ax = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    errorbar=None,
    ax=ax
)

ax.set_title("Average Total Bill by Day and Sex")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Total Bill ($)")

ax.legend(title="Sex")
plt.tight_layout()

plt.show()
```
