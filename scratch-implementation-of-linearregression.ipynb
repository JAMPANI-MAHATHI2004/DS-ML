Linear Regression with Gradient Descent - for large data

```python
# Creating Linear Regression
class LinearRegression:
    def __init__(self, learning_rate = 0.01, n_iter=1000):# we will create contructor as we have few values to initialize
        self.bias = None # we use self beacause we have to make them(bias, weights) parameters for har ek individual not just class
        self.weights = None
        self.lr = learning_rate
        self.n_iter = n_iter

    def fit(self, X, y): # X_train, y_train
        m, n = X.shape # (number of samples, number of features)
        
        # step 1 - Initialize the parameters
        self.bias = 0
        self.weights = np.zeros(n) #[0,0,0,...0] 
        
        # this loop represents Gradient Descent
        for i in range(self.n_iter):
            # step 2 - Calculate y_pred
            y_pred = self.bias + np.dot(X, self.weights)
            
            # step 3 - Calculate Gradient
            db = (1/m) * np.sum(y_pred - y) # derivatives with respect to bias
            dw = (1/m) * np.dot(X.T, (y_pred - y)) # derivatives with respect to weights
            
            # step 4 - Apply Convergence Theorem and then Update Parameters
            self.bias -= self.lr * db
            self.weights -= self.lr * dw
        
    def predict(self, X):
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
print(y_pred)
print(model.bias)
print(model.weights)
```

Linear Regression with Ordinary Least Squares(OLS) - works for little data

```python
class LinearRegressionOLS:
    def __init__(self):
        self.bias = None
        self.weights = None

    def fit(self, X, y):
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]
        # normal equation
        # theta = np.dot(np.linealg.inv(np.dot(X_b.T, X_b)), np.dot(X_b.T, y))
        theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y   # @ is shortcut for dot product

        self.bias = theta[0]
        self.weights = theta[1:]

    def predict(self, X):
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegressionOLS()
model.fit(X, y)

y_pred = model.predict(X)
print(y_pred)
print(model.bias)
print(model.weights)
```
