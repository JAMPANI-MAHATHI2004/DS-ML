```python
import json
```

```python
# loading the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
load_data("store_data.json")
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

1. data is inconsistant -> mixed type(rating)
2. missing data(age)
3. Duplicate data  
these are the things that we found in our data

```python
# clean and structure thr data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        # clean ratings
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating
        
        # handling missing values
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
            
        # deduplication
        if(user["name"].strip() in unique_users):
            continue
        cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```
